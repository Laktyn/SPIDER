# Does fringes period depend from the system?

```python
# no fiber

phase_spectrum = pd.read_csv("data/no_fiber_shear.csv", skiprows = 2)
temporal_spectrum = pd.read_csv("data/no_fiber_no_shear.csv", skiprows = 2)

phase_spectrum = sa.wl_to_freq(phase_spectrum)
temporal_spectrum = sa.wl_to_freq(temporal_spectrum)

phase_spectrum = sa.constant_spacing(phase_spectrum)
temporal_spectrum = sa.constant_spacing(temporal_spectrum)

# fiber of 20 m

phase_spectrum_20 = pd.read_csv("data/fiber20_shear.csv", skiprows = 2)
temporal_spectrum_20 = pd.read_csv("data/fiber20_no_shear.csv", skiprows = 2)

phase_spectrum_20 = sa.wl_to_freq(phase_spectrum_20)
temporal_spectrum_20 = sa.wl_to_freq(temporal_spectrum_20)

phase_spectrum_20 = sa.constant_spacing(phase_spectrum_20)
temporal_spectrum_20 = sa.constant_spacing(temporal_spectrum_20)

# fiber of 50 m

phase_spectrum_50 = pd.read_csv("data/fiber50_shear.csv", skiprows = 2)
temporal_spectrum_50 = pd.read_csv("data/fiber50_no_shear.csv", skiprows = 2)

phase_spectrum_50 = sa.wl_to_freq(phase_spectrum_50)
temporal_spectrum_50 = sa.wl_to_freq(temporal_spectrum_50)

phase_spectrum_50 = sa.constant_spacing(phase_spectrum_50)
temporal_spectrum_50 = sa.constant_spacing(temporal_spectrum_50)

# fiber of 100 m

phase_spectrum_100 = pd.read_csv("data/fiber100_shear.csv", skiprows = 2)
temporal_spectrum_100 = pd.read_csv("data/fiber100_no_shear.csv", skiprows = 2)

phase_spectrum_100 = sa.wl_to_freq(phase_spectrum_100)
temporal_spectrum_100 = sa.wl_to_freq(temporal_spectrum_100)

phase_spectrum_100 = sa.constant_spacing(phase_spectrum_100)
temporal_spectrum_100 = sa.constant_spacing(temporal_spectrum_100)

# fiber of 150 m

phase_spectrum_150 = pd.read_csv("data/fiber150_shear.csv", skiprows = 2)
temporal_spectrum_150 = pd.read_csv("data/fiber150_no_shear.csv", skiprows = 2)

phase_spectrum_150 = sa.wl_to_freq(phase_spectrum_150)
temporal_spectrum_150 = sa.wl_to_freq(temporal_spectrum_150)

phase_spectrum_150 = sa.constant_spacing(phase_spectrum_150)
temporal_spectrum_150 = sa.constant_spacing(temporal_spectrum_150)

# fiber of 250 m

phase_spectrum_250 = pd.read_csv("data/fiber250_shear.csv", skiprows = 2)
temporal_spectrum_250 = pd.read_csv("data/fiber250_no_shear.csv", skiprows = 2)

phase_spectrum_250 = sa.wl_to_freq(phase_spectrum_250)
temporal_spectrum_250 = sa.wl_to_freq(temporal_spectrum_250)

phase_spectrum_250 = sa.constant_spacing(phase_spectrum_250)
temporal_spectrum_250 = sa.constant_spacing(temporal_spectrum_250)

# and plot it

start = 191.8
end = 192.5

sa.compare_plots([phase_spectrum, temporal_spectrum], title = "No fiber", start = start, end = end, legend = ["Sheared", "Not sheared"])
print("[EOPM off] Period of fringes equal to {} +- {} nm".format(round(sa.find_period(temporal_spectrum)[0], 4), round(sa.find_period(temporal_spectrum)[1], 4)))
print("[EOPM on] Period of fringes equal to {} +- {} nm".format(round(sa.find_period(phase_spectrum)[0], 4), round(sa.find_period(phase_spectrum)[1], 4)))

sa.compare_plots([phase_spectrum_20, temporal_spectrum_20], title = "50 m fiber", start = start, end = end, legend = ["Sheared", "Not sheared"])
print("[EOPM off] Period of fringes equal to {} +- {} nm".format(round(sa.find_period(temporal_spectrum_20)[0], 4), round(sa.find_period(temporal_spectrum_20)[1], 4)))
print("[EOPM on] Period of fringes equal to {} +- {} nm".format(round(sa.find_period(phase_spectrum_20)[0], 4), round(sa.find_period(phase_spectrum_20)[1], 4)))

sa.compare_plots([phase_spectrum_50, temporal_spectrum_50], title = "50 m fiber", start = start, end = end, legend = ["Sheared", "Not sheared"])
print("[EOPM off] Period of fringes equal to {} +- {} nm".format(round(sa.find_period(temporal_spectrum_50)[0], 4), round(sa.find_period(temporal_spectrum_50)[1], 4)))
print("[EOPM on] Period of fringes equal to {} +- {} nm".format(round(sa.find_period(phase_spectrum_50)[0], 4), round(sa.find_period(phase_spectrum_50)[1], 4)))

sa.compare_plots([phase_spectrum_100, temporal_spectrum_100], title = "100 m fiber", start = start, end = end, legend = ["Sheared", "Not sheared"])
print("[EOPM off] Period of fringes equal to {} +- {} nm".format(round(sa.find_period(temporal_spectrum_100)[0], 4), round(sa.find_period(temporal_spectrum_100)[1], 4)))
print("[EOPM on] Period of fringes equal to {} +- {} nm".format(round(sa.find_period(phase_spectrum_100)[0], 4), round(sa.find_period(phase_spectrum_100)[1], 4)))

sa.compare_plots([phase_spectrum_150, temporal_spectrum_150], title = "150 m fiber", start = start, end = end, legend = ["Sheared", "Not sheared"])
print("[EOPM off] Period of fringes equal to {} +- {} nm".format(round(sa.find_period(temporal_spectrum_150)[0], 4), round(sa.find_period(temporal_spectrum_150)[1], 4)))
print("[EOPM on] Period of fringes equal to {} +- {} nm".format(round(sa.find_period(phase_spectrum_150)[0], 4), round(sa.find_period(phase_spectrum_150)[1], 4)))

start = 191.75
end = 192.45

sa.compare_plots([phase_spectrum_250, temporal_spectrum_250], title = "250 m fiber", start = start, end = end, legend = ["Sheared", "Not sheared"])
print("[EOPM off] Period of fringes equal to {} +- {} nm".format(round(sa.find_period(temporal_spectrum_250)[0], 4), round(sa.find_period(temporal_spectrum_250)[1], 4)))
print("[EOPM on] Period of fringes equal to {} +- {} nm".format(round(sa.find_period(phase_spectrum_250)[0], 4), round(sa.find_period(phase_spectrum_250)[1], 4)))
```
