```python
import spectral_analysis as sa
import numpy as np
import matplotlib.pyplot as plt
```

```python
spectre = sa.hermitian_pulse([190, 196], 193, 1, "freq")
sa.plot(spectre, what_to_plot = "real")
spectre.zero_padding(5)
spectre.fourier()
for i in range(len(spectre)):
    if i%2 == 0:
        spectre.Y[i] *= -1
sa.plot(spectre, what_to_plot = "real", start = -5, end = 5)
```

```python
spectre = sa.gaussian_pulse([190, 196], 193, 1, "freq")
X = spectre.X.copy()
Y = spectre.Y.copy()

from scipy.fft import fft, fftshift

plt.plot(X, np.real(Y), color = "red")
plt.plot(X, np.imag(Y), color = "green")
plt.legend(["Real part", "Imaginary part"])
plt.title("Before FFT")
plt.show()

Y = fft(Y)
Y = fftshift(Y)

for i in range(len(X)):
    break
    if i%2 == 0:
        Y[i] *= -1

plt.plot(X[980:1020], np.real(Y[980:1020]), color = "red")
plt.plot(X[980:1020], np.imag(Y[980:1020]), color = "green")
plt.legend(["Real part", "Imaginary part"])
plt.title("After FFT")
plt.show()
```
