# Old measurements

```python
m80_extracted_smoothed = sa.load_csv("80m_extracted_smoothed_2.csv", x_type = "freq", y_type = "phase")
m80_extracted_not_smoothed = sa.load_csv("80m_extracted_not_smoothed.csv", x_type = "freq", y_type = "phase")
m40_extracted_smoothed = sa.load_csv("40m_extracted_smoothed.csv", x_type = "freq", y_type = "phase")
m40_measured_smoothed = sa.load_csv("40m_measured_smoothed.csv", x_type = "freq", y_type = "phase")
simulation_40m_1 = sa.load_csv("simulation_40m_1.csv", x_type = "freq", y_type = "phase")
simulation_80m = sa.load_csv("simulation_80m.csv", x_type = "freq", y_type = "phase")

m80_extracted_smoothed.shift(-0.37)

sa.compare_plots([m80_extracted_smoothed, 
        m80_extracted_not_smoothed, 
        m40_extracted_smoothed,
        m40_measured_smoothed,
        simulation_40m_1,
        simulation_80m],
        colors = ["red", "orange", "lime", "blue", "green", "brown"],
        title = "Spectral phases for different setups",
        legend = ["80m, extracted shear, smoothed (v1)",
                "80m, extracted shear, not_smoothed (v2)",
                "40m, extracted shear, smoothed (v2)",
                "40m, measured shear, smoothed (v2)",
                "40m, simulation",
                "80m, simulation"],)
```
